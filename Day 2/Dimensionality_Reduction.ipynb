```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
sns.set()

# Import and load the diabetes dataset
X, Y = load_diabetes(return_X_y=True)
```

The diabetes dataset contains the following 11 features. The 11th feature is the target variable. This is a regression task.
1. Age
2. Sex
3. Body mass index
4. Average blood pressure
5. S1
6. S2
7. S3
8. S4
9. S5
10. S6
11. Quantitative measure of disease progression one year after baseline

```python
# A simple but naive method of visualizing this high dimensional data is to plot the first/any two features/dimension
plt.scatter(X[:,2], X[:,3], c=Y)
plt.title('Body Mass Index vs Average Blood Pressure')
plt.xlabel('Body Mass Index')
plt.ylabel('Average Blood Pressure')
plt.show()
# Darker colours corresspond to larger values
```

```python
# A better method is to perform dimensionality reduction on the dataset
# PCA gives the first n principal components (n eigen vectors corressponding to the n largest eigen values)

# create the PCA instance
pca = PCA(2)
# fit on data and get the new projection
X_low = pca.fit_transform(X)

# Now let's plot the reduced dataset
plt.scatter(X_low[:,0], X_low[:,1], c=Y)
plt.title('PC1 vs PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
